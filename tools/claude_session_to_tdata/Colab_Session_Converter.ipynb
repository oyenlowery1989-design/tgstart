# 🔄 Telethon Session to TDesktop Converter (Colab Edition)

This notebook allows you to convert your `.session` files to Telegram Desktop `tdata` format directly in the cloud. This avoids local environment issues on Windows.

### 🚀 Instructions:
1. Run the **Installation** cell.
2. Run the **Upload** cell to upload your `.session` file.
3. Run the **Configuration** cell and enter your API ID and API Hash.
4. Run the **Conversion** cell.
5. Download the resulting `tdata.zip` file.

```python
# @title 1. Install Dependencies
!pip install -q telethon opentele tgcrypto
```

```python
# @title 2. Upload .session File
from google.colab import files
import os

uploaded = files.upload()
session_file = list(uploaded.keys())[0]
print(f"Uploaded: {session_file}")
```

```python
# @title 3. Configuration
import getpass

API_ID = input("Enter API ID: ")
API_HASH = getpass.getpass("Enter API Hash: ")
SESSION_NAME = session_file.replace(".session", "")
```

```python
# @title 4. Convert Session
import asyncio
from opentele.td import TDesktop
from opentele.tl import TelegramClient
from opentele.api import API, UseCurrentSession

async def convert():
    client = TelegramClient(SESSION_NAME, int(API_ID), API_HASH)
    await client.connect()
    
    if not await client.is_user_authorized():
       print("❌ Error: Session is not authorized!")
       return
    
    td = await client.ToTDesktop(flag=UseCurrentSession, api=API.TelegramDesktop)
    td.SaveTData("./tdata")
    print("✅ Success: tdata generated in ./tdata folder")
    await client.disconnect()

await convert()
```

```python
# @title 5. Download tdata
import shutil
shutil.make_archive('tdata_output', 'zip', './tdata')
files.download('tdata_output.zip')
```
